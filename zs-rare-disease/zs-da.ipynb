```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
train_file = pd.read_csv('train.csv')
train_file.head()
```

```python
labels = train_file['Disease Flag']
feature_vars = ['var' + str(i) for i in range(1, 27)]
features = train_file[feature_vars]
features.shape
```

```python
from collections import Counter
train_file['Disease Flag'].value_counts()
train_file['var26'].head()
```

```python
%matplotlib inline
from sklearn import preprocessing

to_normalize_vars = ['var1', 'var2', 'var4', 'var5', 'var9', 'var10', 'var11', 'var15',
                     'var17', 'var18', 'var20', 'var21', 'var22']

for var in to_normalize_vars:
    x = train_file[var].values.astype(float)
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    train_file[var] = x_scaled
train_file.head()
```

```python
from sklearn.feature_selection import mutual_info_regression

mi = mutual_info_regression(train_file[feature_vars], labels)
mi
```

```python
# enumerate(mi)
# from operator import itemgetter
# print(sorted(enumerate(mi), key=itemgetter(1)))
feature_vars = ['var' + str(i + 1) for i in [22, 23, 5, 17, 14, 15, 12, 0, 4, 1, 13, 11, 2]]
```

```python
# Do feature selection
# Try other models like linearSVC, DecisionTree, Neural Networks etc

from sklearn.linear_model import LogisticRegression
from sklearn import svm  # dont use
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier, MLPRegressor

# lg_model = svm.SVC(probability=1, class_weight='balanced', verbose=True)
lg_model = MLPRegressor(activation='logistic')
lg_model.fit(train_file[feature_vars], labels)
```

```python
test_file = pd.read_csv('test.csv')
test_file.head()

for var in to_normalize_vars:
    x = test_file[var].values.astype(float)
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    test_file[var] = x_scaled
test_file.head()
```

```python
prediction = lg_model.predict(test_file[feature_vars])
```

```python
output = pd.DataFrame()
output['Resident ID'] = test_file['Resident ID']
output['Disease Flag'] = prediction
output.to_csv('sample.csv', index=False)
```
